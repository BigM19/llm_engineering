To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os 
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display, update_display
```

```python
# constants
MODEL_GEMINI = 'gemini-2.5-flash'
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
MODEL_LLAMA = 'llama3.2:1b'
OLLAMA_BASE_URL = "http://localhost:11434/v1"
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv("GEMINI_API_KEY")

if api_key and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please visit the troubleshooting notebook!")
    
gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=api_key)
llama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

```python
system_prompt = """
    You are a professional Python programmer. You are helping a junior programmer to write and understand Python code.
    You will be given a task, either to explain what a piece of code does, or to write a piece of code, or even correct
    it if it contains errors.
    You should respond with the answer in markdown format, without code blocks.
"""
```

```python
# here is the question; type over this to ask something new
def create_question(code_snippet):
    user_prompt = f"""
    Please explain what this code does and why. Explain everything in detail, as if you were explaining it to a beginner.
    Here is the code:
    {code_snippet}
    """
    return user_prompt
```

```python
# Get gemini-2.5-flash to answer, with streaming
input_code = input("Type your code snippet here: ")
stream = gemini.chat.completions.create(
    model=MODEL_GEMINI,
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": create_question(input_code)},
    ],
    stream=True
)
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content
    update_display(Markdown(response), display_id=display_handle.display_id)
```
